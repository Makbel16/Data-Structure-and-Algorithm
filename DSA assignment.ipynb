### **1. Bubble Sort Algorithm using ID <BR>**
##### üîπ Description & Implementation Steps
**Description**:<br>
 - *Bubble Sort* : is a simple comparison-based sorting algorithm that repeatedly steps through a<br> list, compares 
adjacent elements, and swaps them if they are in the wrong order. This <br>process continues until the entire list is sorted.          

**Given Data**:<br>
- A list of 10 students, where each student is represented as a dictionary with id, name, dept, and cgpa.<br>   

**Objective**:<br>         
- Sort the students by ID in ascending order using Bubble Sort<br>          
##### üîπ **Implementation Steps**:<br>
 **Step 1**
  - Define the Student Data Prepare the data by creating a list of student records. Each record includes id, name, dept, and cgpa.<br>

**Step 2**:
- Apply Bubble Sort Go through the list multiple times:<br>  
- In each pass, compare two adjacent student records.<br>  
- If the student on the left has a higher ID than the one on the right, swap them.<br>  
- Repeat this for the entire list, reducing the number of comparisons each time.<br>  
- Continue this process until no swaps are needed ‚Äî indicating the list is sorted.<br>  

**Step** 3: 
- Display Sorted Results<br> 
Once sorting is complete, the student records will be arranged in ascending order by their ID ‚Äî from the smallest
- (e.g., "DBU1500001") to the largest (e.g., "DBU1500010").<br>  


```python
# Original unsorted student list
Students = [
    {'id': 'DBU1500001', 'name': 'Dawit', 'dept': 'DS', 'cgpa': 3.34},
    {'id': 'DBU1500005', 'name': 'Abrham', 'dept': 'CS', 'cgpa': 3.12},
    {'id': 'DBU1500003', 'name': 'Soliana', 'dept': 'DS', 'cgpa': 2.74},
    {'id': 'DBU1500002', 'name': 'Rosa', 'dept': 'CS', 'cgpa': 3.10},
    {'id': 'DBU1500009', 'name': 'Petros', 'dept': 'SE', 'cgpa': 2.90},
    {'id': 'DBU1500007', 'name': 'Yosef', 'dept': 'IT', 'cgpa': 3.67},
    {'id': 'DBU1500006', 'name': 'Birhanu', 'dept': 'CS', 'cgpa': 2.59},
    {'id': 'DBU1500008', 'name': 'Makbel', 'dept': 'CS', 'cgpa': 3.96},
    {'id': 'DBU1500010', 'name': 'Kalkidan', 'dept': 'SE', 'cgpa': 3.80},
    {'id': 'DBU1500004', 'name': 'Nahom', 'dept': 'CS', 'cgpa': 3.00}
]

def bubble_sort_by_id(student_list):
    n = len(student_list)
    for i in range(n):
        for j in range(0, n - i - 1):
            if student_list[j]['id'] > student_list[j + 1]['id']:
                student_list[j], student_list[j + 1] = student_list[j + 1], student_list[j]
    return student_list

sorted_by_id = bubble_sort_by_id(Students.copy())

print("Sorted by ID (Bubble Sort):")
print("[")
for s in sorted_by_id:
    print(f"    {s},")
print("]")
```

### **2. Insertion Sort (by Name)**
### üîπ Description

- **Insertion Sort** is a simple and intuitive sorting algorithm that builds the final sorted list one item at a time. It is similar to how we sort playing cards in our hands. The algorithm assumes that the first element is already sorted, and it inserts each next element into the correct position relative to the sorted portion of the list.

In this implementation, the students are sorted **alphabetically by their names** using the `Insertion Sort` algorithm.



### üîπ Step-by-Step Process

- **Step 1**. Start from the second element (index 1) because a single element (index 0) is already considered sorted.     

    - **Step 2**. Pick the current element (`key`) and compare it with elements in the sorted part of the list (left side).

    - **Step 3**. **Shift all elements that are **greater than the key** to the right to make space for inserting the key.

    - **Step 4**. **Insert the key at its correct position in the sorted section.

    - **Step 5**. Repeat steps 2‚Äì4 until the whole list is sorted.






```python
Students = [
    {'id': 'DBU1500001', 'name': 'Dawit', 'dept': 'DS', 'cgpa': 3.34},
    {'id': 'DBU1500005', 'name': 'Abrham', 'dept': 'CS', 'cgpa': 3.12},
    {'id': 'DBU1500003', 'name': 'Soliana', 'dept': 'DS', 'cgpa': 2.74},
    {'id': 'DBU1500002', 'name': 'Rosa', 'dept': 'CS', 'cgpa': 3.10},
    {'id': 'DBU1500009', 'name': 'Petros', 'dept': 'SE', 'cgpa': 2.90},
    {'id': 'DBU1500007', 'name': 'Yosef', 'dept': 'IT', 'cgpa': 3.67},
    {'id': 'DBU1500006', 'name': 'Birhanu', 'dept': 'CS', 'cgpa': 2.59},
    {'id': 'DBU1500008', 'name': 'Makbel', 'dept': 'CS', 'cgpa': 3.96},
    {'id': 'DBU1500010', 'name': 'Kalkidan', 'dept': 'SE', 'cgpa': 3.80},
    {'id': 'DBU1500004', 'name': 'Nahom', 'dept': 'CS', 'cgpa': 3.00}
]


def insertion_sort_by_name(student_list):
    for i in range(1, len(student_list)):
        key = student_list[i]
        j = i - 1
        while j >= 0 and key['name'].lower() < student_list[j]['name'].lower():
            student_list[j + 1] = student_list[j]
            j -= 1
        student_list[j + 1] = key
    return student_list

sorted_by_name = insertion_sort_by_name(Students.copy())
print("\nSorted by Name (Insertion Sort):")
for student in sorted_by_name:
    print(student)
```

   ### **3.Selection Sort Algorithm using CGPA (Descending Order) <br>**
 **Description & Implementation Steps**<br>
‚û°Ô∏è **Description**:<br>
- *Selection Sort* : is a comparison-based sorting algorithm. It works by dividing the list into two parts: the sorted part (initially empty) and the unsorted part. The algorithm repeatedly selects the maximum value from the unsorted part and moves it to the beginning of the sorted part.

- In this case, we use Selection Sort to arrange students in descending order of their CGPA, meaning the student with the highest CGPA comes first, and the one with the lowest CGPA comes last. 

‚û°Ô∏è **Objective**:<br>
- Sort the list of students by CGPA in descending order using the Selection Sort algorithm.

‚û°Ô∏è **Implementation Steps**:<br>
  - Step 1: Start with the unsorted list of students.<br>  

  - Step 2: For each position i in the list (from 0 to n-1): Assume the current position i holds the student with the highest CGPA in the remaining list. Compare it with the rest of the list (from i+1 to end). and also If any student has a higher CGPA, update the max_idx to that student's index.<br>

  - Step 3: Swap the student at i with the student at max_idx.<br> This places the highest CGPA student in the correct position.<br> 

  - Step 4: Repeat until the whole list is sorted in descending CGPA order.<br>

```python
Students = [
    {'id': 'DBU1500001', 'name': 'Dawit', 'dept': 'DS', 'cgpa': 3.92},
    {'id': 'DBU1500005', 'name': 'Abrham', 'dept': 'CS', 'cgpa': 3.12},
    {'id': 'DBU1500003', 'name': 'Soliana', 'dept': 'DS', 'cgpa': 2.74},
    {'id': 'DBU1500002', 'name': 'Rosa', 'dept': 'CS', 'cgpa': 3.10},
    {'id': 'DBU1500009', 'name': 'Petros', 'dept': 'SE', 'cgpa': 2.90},
    {'id': 'DBU1500007', 'name': 'Yosef', 'dept': 'IT', 'cgpa': 3.67},
    {'id': 'DBU1500006', 'name': 'Birhanu', 'dept': 'CS', 'cgpa': 2.59},
    {'id': 'DBU1500008', 'name': 'Makble', 'dept': 'CS', 'cgpa': 3.96},
    {'id': 'DBU1500010', 'name': 'Kalkidan', 'dept': 'SE', 'cgpa': 3.80},
    {'id': 'DBU1500004', 'name': 'Nahom', 'dept': 'CS', 'cgpa': 3.00}
]
def selection_sort_by_cgpa_desc(student_list):
    n = len(student_list)
    for i in range(n):
        max_idx = i
        for j in range(i+1, n):
            if student_list[j]['cgpa'] > student_list[max_idx]['cgpa']:
                max_idx = j
        student_list[i], student_list[max_idx] = student_list[max_idx], student_list[i]
    return student_list

sorted_by_cgpa = selection_sort_by_cgpa_desc(Students.copy())

print("\nSorted by CGPA Descending (Selection Sort):")
print("[")
for i in sorted_by_cgpa:
    print(f"   {i},")
print("]")
```

## üîπComparative Analysis of Sorting Algorithms

This section presents a detailed comparison of **Bubble Sort**, **Insertion Sort**, and **Selection Sort** based on the earlier Python implementations used to sort student profiles.


### - Overview Table      

| **Algorithm**     | **Best Case**      | **Average Case**    | **Worst Case**       | **Time Complexity** | **Space Complexity** | **Stable** | **Adaptive** | **Use Case Suitability**                |
|-------------------|--------------------|----------------------|------------------------|----------------------|------------------------|------------|--------------|-----------------------------------------|
| **Bubble Sort**   | O(n)               | O(n¬≤)                | O(n¬≤)                  | O(n¬≤)               | O(1)                   |  Yes     |  Yes       | Small & nearly sorted datasets          |
| **Insertion Sort**| O(n)               | O(n¬≤)                | O(n¬≤)                  | O(n¬≤)               | O(1)                   |  Yes     |  Yes       | Small or partially sorted data          |
| **Selection Sort**| O(n¬≤)              | O(n¬≤)                | O(n¬≤)                  | O(n¬≤)               | O(1)                   |  No      |  No        | Simple logic, fewer swaps needed        |




###  Detailed Descriptions

### üîπ **Bubble Sort**
- **Best Case (O(n))**: When the list is already sorted. Only one pass is needed with no swaps.
- **Worst Case (O(n¬≤))**: When the list is in reverse order. Every element must be compared and swapped.
- **Average Case (O(n¬≤))**: In most cases, many swaps and comparisons are required.
- **Use Case**: Best for very small lists or for learning purposes.
- **Stable & Adaptive**<br>

  **Advantages**:  
  - Simple and easy to implement. 
  - Stable sorting (keeps relative order of equal elements). 
  - Adaptive ‚Äî performs well when list is nearly sorted.     

  **Disadvantages**:
  - Very inefficient for large datasets.
  - Many unnecessary swaps slow it down.  

---

### üîπ **Insertion Sort**
- **Best Case (O(n))**: The list is already sorted; only one comparison per element.
- **Worst Case (O(n¬≤))**: The list is in reverse order; each element is compared to all previous ones.
- **Average Case (O(n¬≤))**: Moderate comparisons and shifting are needed.
- **Use Case**: Great for small datasets or when the data is nearly sorted.
- **Stable & Adaptive**

  **Advantages**:  
  - Efficient for small or nearly sorted datasets.
  - Stable sorting algorithm.
  - Adaptive ‚Äî fewer comparisons when data is mostly sorted.
  - Good for real-time data processing.  

  **Disadvantages**:
  - Inefficient for large unsorted data.
  - Still O(n¬≤) for average and worst cases.   

---

### üîπ **Selection Sort**
- **Best = Average = Worst Case (O(n¬≤))**: Because it always goes through the unsorted part of the list to find the minimum (or maximum), no matter the initial order.
- **Use Case**: Suitable when you want to minimize the number of swaps.
- Not Stable, Not Adaptive.     

  **Advantages**:
  - Simple logic and fewer swaps than Bubble Sort.
  - Useful when memory writes/swaps are expensive.

  **Disadvantages**:
  - Not stable ‚Äî equal items may change order.
  - Not adaptive ‚Äî doesn‚Äôt benefit from partial sorting.
  - Always performs O(n¬≤) comparisons.
